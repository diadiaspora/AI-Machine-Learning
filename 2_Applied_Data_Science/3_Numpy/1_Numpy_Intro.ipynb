__arrays strides__

how many bytes you have to skip in memory to move to the next element along each axis of an array. It tells NumPy how the array is laid out in memory.

```python
arr2d.strides
```

```python
print(arr2d.strides)
```

Strides = memory step size

```python
import numpy as np

arry = np.array([10, 20, 30, 40], dtype=np.int32)
```

Saved in memory like this: 
| 10 | 20 | 30 | 40 |

Each int32 = 4 bytes

```python
arry.strides

```

To move from one element to the next → move 4 bytes

1D array → only one axis (axis 0)

Each element = 4 bytes

To move to the next element → move 4 bytes

| Axis   | Meaning |
| ------ | ------- |
| axis 0 | rows    |
| axis 1 | columns |


Strides are listed as:

(row stride, column stride)



```python
arry2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
], dtype=np.int32)
```

```python
arry2.shape
# (2, 3)
```

Row 0 → 1  2  3

Row 1 → 4  5  6


```python
arry2.strides
```

| Axis | Meaning             | Bytes    |
| ---- | ------------------- | -------- |
| 0    | Move down 1 row     | 12 bytes |
| 1    | Move right 1 column | 4 bytes  |


1 row = 3 elements × 4 bytes = 12 bytes

1 column = 1 element × 4 bytes = 4 bytes

### __`linespace()`__

creates an ndarray of evenly spaced numbers within a defined interval.

The num argument determines how many elements will be in the array, and the function calculates the intervening numbers so that the intervals between them are the same.

```python
np.linspace(start=0, stop=20, num=6)
```

a 1D array of six floating-point values, with all the values evenly spaced. 

force the function to not include the endpoint by setting the Boolean parameter endpoint to False:

```python
np.linspace(0, 20, 6, endpoint=False)
```

retrieve the size of the intervals between values, set the Boolean parameter retstep to True. 

```python
arr1d, step = np.linspace(0, 20, 6, retstep=True)
```

```python
print(arr1d)
```

```python
step
```

By default, the linspace() function returns a dtype of float64. You can override this by passing it a dtype argument:

```python
np.linspace(0, 20, 6, dtype='int64')
```

```python
np.linspace(0, 20, 6).reshape(2, 3)
```

```python
np.eye(N=3, M=3, k=0)
```

It creates a 2D array (matrix) with 1s on a diagonal and 0s everywhere else.

N=3

* Number of rows

M=3

* Number of columns

k=0

* Which diagonal gets the 1s

What does k mean?

Think of diagonals relative to the main diagonal:

| `k` value | Diagonal                    |
| --------- | --------------------------- |
| `0`       | Main diagonal               |
| `1`       | One above the main diagonal |
| `-1`      | One below the main diagonal |


```python
np.eye(3, 3, k=1)
```

```python
np.eye(3, 3, k=-1)
```
